## Quick Solanke and Bournemouth cumulative goals analysis

This is for the 23/24 season – Bournemouth's first under Iraola.

This quick and dirty analysis shows:

1. Cumulative goals
2. Cumulative goals with trend line
3. Cumulative xG
4. Solanke's cumulative xG
5. Solanke’s xG per game for 23-24 with range-specific averages

It shows how something changed for Bournemouth around matchweek 10 to 12.

**Matchweek 10** was Iraola's first victory of the season; a 2-1 win at home against Burnley;

**Matchweek 11** was a devastating 1-6 loss away at City;

**Matchweek 12** was the watershed moment: an emphatic 2-0 home win against Newcastle.

Following that, they went on an impressive unbeaten streak, more than doubling their goal output from 8 in the first ten games to 20 in the next ten.

Their average goals per game went from 0.8 in the first ten matches to 1.5 for the rest of the season. That's an 87.5% increase in output.

**Something clicked.**

Original data source: FBref/Opta

## 1. Bournemouth's cumulative goals 23-24

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager

font_dirs = ["/usr/share/fonts/truetype/roboto"]  # The path to the custom font file.
font_files = matplotlib.font_manager.findSystemFonts(fontpaths=font_dirs)

for font_file in font_files:
    matplotlib.font_manager.fontManager.addfont(font_file)

# Data
data = [1, 1, 0, 2, 0, 1, 0, 0, 1, 2, 1, 2, 3, 2, 2, 3, 3, 3, 1, 0, 1, 1, 1, 2, 0, 2, 2, 4, 1, 1, 1, 2, 1, 1, 3, 0, 1, 0]

df = pd.DataFrame(data, columns=['gls'])

# Calculating cumulative total
df['Cumulative_Total'] = df['gls'].cumsum()

# Set default graph resolution
plt.rcParams['figure.dpi'] = 150
plt.rcParams['font.family'] = ['Roboto', 'sans-serif']

# Plotting the cumulative total
plt.figure(figsize=(10, 6))
plt.plot(df['Cumulative_Total'], marker='o')

# Titles and labels
plt.axvline(x=11, color='red', linestyle='--', label='12th Tick Marker')
plt.title('Bournemouth\'s cumulative goals 23-24')
plt.xlabel('Index')
plt.ylabel('Cumulative Total')
plt.grid(True)

plt.show()
```

## 2. Bournemouth’s cumulative goals with trend line

```python
x = np.arange(len(df['Cumulative_Total']))
z_granular = np.polyfit(x, df['Cumulative_Total'], 4)
p_granular = np.poly1d(z_granular)

# Plotting the cumulative total and trendline
plt.figure(figsize=(10, 6))
plt.plot(df['Cumulative_Total'], marker='o', label='Cumulative Total')

# Add trendline
plt.plot(x, p_granular(x), color='gray', linestyle='--', label='Granular Trend Line')

# Titles and labels
plt.title('Bournemouth\'s cumulative goals with trend line')
plt.xlabel('Index')
plt.ylabel('Cumulative Total')
plt.axvline(x=11, color='red', linestyle='--', label='12th Tick Marker')
plt.grid(True)
plt.legend()

plt.show()
```

## 3. Bournemouth’s cumulative xG

```python
xg_data = [1.3, 1.3, 0.7, 2.1, 1.0, 1.1, 0.6, 0.7, 0.6, 1.1, 0.9, 2.4, 2.9, 2.0, 1.8, 1.5, 1.6, 1.8, 1.7, 1.4, 1.2, 
           1.3, 1.8, 2.2, 0.9, 1.1, 3.3, 2.3, 1.3, 1.1, 0.4, 1.6, 1.8, 1.9, 1.9, 0.5, 0.7, 2.2]

df_xg = pd.DataFrame(xg_data, columns=['xG'])

# Calculate cumulative total for xG
df_xg['Cumulative_Total_xG'] = df_xg['xG'].cumsum()

# Plot the cumulative total of xG
plt.figure(figsize=(10, 6))
plt.plot(df_xg['Cumulative_Total_xG'], marker='o', label='Cumulative Total xG')

# Titles and labels
plt.title('Bournemouth\'s cumulative xG')
plt.xlabel('Index')
plt.ylabel('Cumulative Total xG')
plt.axvline(x=11, color='red', linestyle='--', label='12th Tick Marker')
plt.grid(True)
plt.legend()

plt.show()
```

## 4. Solanke’s cumulative xG for 23-24

```python
# Solanke's xG 23-24 season
solanke_xg_data = [1, 0.1, 0.1, 0.6, 0.4, 0.5, 0.2, 0.1, 0.5, 0.2, 0.6, 1.1, 0.6, 1.3, 0.2, 0.9, 0.7, 1, 0.8, 0.1, 
                   0.5, 0, 0.7, 1.9, 0.3, 0.2, 0.9, 0.4, 0.6, 0.1, 0.1, 0.3, 0.8, 0.1, 0.3, 0.2, 0.3, 0.5]

# Creating DataFrame for Solanke's xG
df_solanke_xg = pd.DataFrame(solanke_xg_data, columns=['xG'])

# Calculating cumulative total for Solanke's xG
df_solanke_xg['Cumulative_Total_xG'] = df_solanke_xg['xG'].cumsum()

# Plotting the cumulative total of Solanke's xG without the trendline
plt.figure(figsize=(10, 6))
plt.plot(df_solanke_xg['Cumulative_Total_xG'], marker='o', label="Cumulative Total xG")

# Titles and labels
plt.title("Solanke's xG for 23-24")
plt.xlabel('Index')
plt.ylabel('Cumulative Total xG')
plt.axvline(x=11, color='red', linestyle='--', label='12th Tick Marker')
plt.grid(True)
# plt.xticks(ticks=range(1, len(solanke_xg_data)))
plt.legend()

plt.show()
```

## 5. Solanke’s xG per game for 23-24 with range-specific averages

This chart shows Solanke's average xG for his first 10 games (red; 0.37 xG per game) and his average xG for his remaining games (green; 0.55 xG per game)

```python
# CAlculate the average xG for games 1 to 10 and games 11 onwards
average_xg_1_to_10 = sum(solanke_xg_data[:10]) / len(solanke_xg_data[:10])
average_xg_11_onwards = sum(solanke_xg_data[10:]) / len(solanke_xg_data[10:])

# Plotting bar chart to show xG per game
plt.figure(figsize=(12, 8), dpi=150)
plt.bar(range(1, len(solanke_xg_data) + 1), solanke_xg_data, color='lightsteelblue')

# Green line limited to games 1 to 10
plt.plot([1, 10], [average_xg_1_to_10, average_xg_1_to_10], color='green', linestyle='--', label=f'Average xG (1-10) ({average_xg_1_to_10:.2f})')

# Red line limited to games 11 onwards
plt.plot([11, len(solanke_xg_data)], [average_xg_11_onwards, average_xg_11_onwards], color='red', linestyle='--', label=f'Average xG (11 onwards) ({average_xg_11_onwards:.2f})')

# Titles and labels
plt.title("Solanke's xG per game for 23-24 with range-specific averages")
plt.xlabel('Game Number')
plt.ylabel('xG')
plt.legend()

plt.xlim(0.5, len(solanke_xg_data) + 0.5)

plt.grid(axis='y', linestyle='--')
plt.show()
```
