```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Data
data = [1, 1, 0, 2, 0, 1, 0, 0, 1, 2, 1, 2, 3, 2, 2, 3, 3, 3, 1, 0, 1, 1, 1, 2, 0, 2, 2, 4, 1, 1, 1, 2, 1, 1, 3, 0, 1, 0]

df = pd.DataFrame(data, columns=['gls'])

# Calculating cumulative total
df['Cumulative_Total'] = df['gls'].cumsum()

# Set default graph resolution
plt.rcParams['figure.dpi'] = 150

# Plotting the cumulative total
plt.figure(figsize=(10, 6))
plt.plot(df['Cumulative_Total'], marker='o')

# Titles and labels
plt.axvline(x=11, color='red', linestyle='--', label='12th Tick Marker')
plt.title('Bournemouth\'s cumulative goals 23-24')
plt.xlabel('Index')
plt.ylabel('Cumulative Total')
plt.grid(True)

plt.show()
```

```python
x = np.arange(len(df['Cumulative_Total']))
z_granular = np.polyfit(x, df['Cumulative_Total'], 4)
p_granular = np.poly1d(z_granular)

# Plotting the cumulative total and trendline
plt.figure(figsize=(10, 6))
plt.plot(df['Cumulative_Total'], marker='o', label='Cumulative Total')

# Add trendline
plt.plot(x, p_granular(x), color='red', linestyle='--', label='Granular Trend Line')

# Titles and labels
plt.title('Bournemouth\'s cumulative goals with trend line')
plt.xlabel('Index')
plt.ylabel('Cumulative Total')
plt.axvline(x=11, color='red', linestyle='--', label='12th Tick Marker')
plt.grid(True)
plt.legend()

plt.show()
```
